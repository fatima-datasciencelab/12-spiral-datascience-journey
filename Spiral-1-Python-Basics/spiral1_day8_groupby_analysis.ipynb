```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load dataset
df = pd.read_csv("netflix_titles.csv")

# Step 3: Bar Plot – Top 10 countries by content count
top_countries = df['country'].value_counts().head(10)

#  Plot
plt.figure(figsize=(10,6))
top_countries.plot(kind='bar', color='skyblue')
plt.title("Top 10 Countries by Netflix Content Count")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
# Step 4: Countplot – Ratings distribution
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='rating', hue='rating', order=df['rating'].value_counts().index[:10], palette='cool', legend=False)

plt.title("Top 10 Rating Categories")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.grid(True)
plt.show()
```

```python
# Step 5: Group by Release Year and Type
year_type_group = df.groupby(['release_year', 'type']).size().reset_index(name='count')

#  Line plot to show Movie vs TV Show trend over years
plt.figure(figsize=(14,6))
sns.lineplot(data=year_type_group, x='release_year', y='count', hue='type', marker='o')
plt.title("Netflix Movies vs TV Shows Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

```python
# Step 6: Group by Country and Type
country_type_group = df.groupby(['country', 'type']).size().reset_index(name='count')
top_country_type = country_type_group[country_type_group['country'].isin(top_countries.index)]

# 📊 Barplot: Countries vs Movie/TV
plt.figure(figsize=(12,6))
sns.barplot(data=top_country_type, x='country', y='count', hue='type')
plt.title("Top Countries by Type of Content (Movies vs TV Shows)")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
