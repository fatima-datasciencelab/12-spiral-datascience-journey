```python
# Libraries import kar rahe hain
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Netflix dataset load kar rahe hain
df = pd.read_csv("netflix_titles.csv")


```

```python
# Sirf 'Movie' type filter kar rahe hain
movies = df[df['type'] == 'Movie']
```

```python

# 'duration' column ko numeric banate hain (e.g. "90 min" â†’ 90)

# First, create a full copy of the filtered data
movies = df[df['type'] == 'Movie'].copy()

movies['duration_mins'] = movies['duration'].str.extract(r'(\d+)').astype(float)
```

```python

# ----------------------------------------------
# ðŸ“Œ 1. Mean, Median, Mode
# Mean: sab values ka average hota hai (Average (sab ka total / count)
# Median: beech wali value (sorted list mein)
# Mode: sabse zyada baar aane wali value
print("Mean:", movies['duration_mins'].mean())
print("Median:", movies['duration_mins'].median())
print("Mode:", movies['duration_mins'].mode()[0])
```

```python

# ----------------------------------------------
# ðŸ“Œ 2. Variance & Standard Deviation
# Variance: values average se kitna vary kar rahi hain
# Standard deviation: variance ka square root hota hai (data kitna spread hai) Spread (values kitna farq karti hain average se)
print("Variance:", movies['duration_mins'].var())
print("Standard Deviation:", movies['duration_mins'].std())
```

```python
# ----------------------------------------------
# ðŸ“Œ 3. Min, Max, Range
# Min: sabse chhoti value
# Max: sabse badi value
# Range: Max - Min (kitna farq hai)
print("Min:", movies['duration_mins'].min())
print("Max:", movies['duration_mins'].max())
print("Range:", movies['duration_mins'].max() - movies['duration_mins'].min())
```

```python
# ----------------------------------------------
# ðŸ“Œ 4. IQR (Interquartile Range)
# IQR: Q3 - Q1 â†’ middle 50% data kitna spread hai
# Q1: 25% data se chhoti values
# Q3: 75% data se chhoti values
Q1 = movies['duration_mins'].quantile(0.25)
Q3 = movies['duration_mins'].quantile(0.75)
IQR = Q3 - Q1
print("Q1 (25%):", Q1)
print("Q3 (75%):", Q3)
print("IQR:", IQR)
```

```python
# ----------------------------------------------
# ðŸ“Œ 5. Boxplot
# Boxplot outliers ko detect karne ke liye use hota hai
# Box ke andar middle 50% data hota hai, line is median
# Bahar wale points outliers hote hain
sns.boxplot(x=movies['duration_mins'])
plt.title("Boxplot of Movie Durations")
plt.xlabel("Duration (mins)")
plt.show()
```

```python
# ----------------------------------------------
# ðŸ“Œ 6. Violin Plot
# Violin plot shape aur distribution dikhata hai (thickness = frequency)
# Ye boxplot ka advanced version hai
sns.violinplot(x=movies['duration_mins'], color='lightgreen')
plt.title("Violin Plot of Movie Durations")
plt.xlabel("Duration (mins)")
plt.show()
```

```python
# ----------------------------------------------
# ðŸ“Œ 7. Skewness & Kurtosis
# Skewness: data ka left ya right tilt
# Kurtosis: data peakness ya flatness
print("Skewness:", movies['duration_mins'].skew())
print("Kurtosis:", movies['duration_mins'].kurtosis())
```
