```python
#Line Plot
# Line plots are used to visualize trends over time (e.g., number of Netflix titles released per year).
# It helps identify if data is increasing, decreasing, or fluctuating over a sequence.

# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load Netflix dataset
df = pd.read_csv("netflix_titles.csv")  # Make sure this CSV file is in same folder

# Step 3: Line plot for number of titles released each year
years = df['release_year'].value_counts().sort_index()  # Count of shows/movies per year
plt.plot(years.index, years.values, marker='o', color='blue')
plt.title("Netflix Titles Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.xticks(rotation=45)  # Better readability
plt.tight_layout()
plt.show()

```

```python
#Histogram
# Histograms are used to show the distribution of a numerical column (e.g., movie durations).
# It tells us how frequently values fall into different ranges (called "bins").
# Useful for identifying skewed data or common value ranges.

# Histogram of duration (if available)
df['duration_mins'] = df['duration'].str.extract('(\d+)').astype(float)
plt.hist(df['duration_mins'].dropna(), bins=10, color='green')
plt.title("Histogram of Movie Durations")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```

```python
#scatter plot
# Scatter plots show the relationship between two numeric variables (e.g., release year vs. rating).
# It helps identify patterns, clusters, or correlations between variables.
# Each dot represents one data point.

# Scatter (Example with fake rating column)
import numpy as np
df['rating_value'] = np.random.randint(1, 10, df.shape[0])

plt.scatter(df['release_year'], df['rating_value'], alpha=0.4)
plt.title("Year vs Rating")
plt.xlabel("Release Year")
plt.ylabel("Fake Rating")
plt.show()
```
